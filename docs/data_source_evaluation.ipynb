# yfinance

```python
import yfinance as yf
t = '600089'
if ((t[0] == '0') or (t[0] == '3')):
    ticker = t + '.SZ'
elif t[0] == '6':
    ticker = t + '.SS'
print('ticker :' + ticker)
```

```python
from datetime import datetime

data = yf.Ticker(ticker)
today = datetime.now().date()

df = data.history(period="1d",
                start=today,
                interval="1m")
```

# yliveticker
