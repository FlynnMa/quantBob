# yfinance
### 可以比较不错的获取分钟日的交易数据，用于分析
但是， yfinance目的不在于提供当天的数据， yfinance的回复
```
yfinance is not intended for use with real-time ticker collection.
Also note that even if you can collect data from yahoo, there might be a delay.
```
具体参考我提的[issue-850](https://github.com/ranaroussi/yfinance/issues/850)

````python
import yfinance as yf
t = '600089'
if ((t[0] == '0') or (t[0] == '3')):
    ticker = t + '.SZ'
elif t[0] == '6':
    ticker = t + '.SS'
print('ticker :' + ticker)
````

````python
from datetime import datetime

data = yf.Ticker(ticker)
today = datetime.now().date()

df = data.history(period="1d",
                start=today,
                interval="1m")
````
